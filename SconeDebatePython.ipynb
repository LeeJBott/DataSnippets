```python
import matplotlib.pyplot as plt

# Sample data
Method = ['Jam then Cream', 'Cream then Jam', 'No Answer', 'Any Combination', 'Half & Half', 'No Scone!', 'Only Jam', 'Only Cream', 'Cheese']
Count = [8, 6, 42, 7, 4, 4, 2, 1, 1]

# Create bar graph
plt.bar(Method, Count)
# Add title and labels
plt.title('The Scone Debate')
plt.xlabel('Preference')
plt.ylabel('Number of Votes')

plt.tick_params(axis='x', labelsize=5)
plt.tick_params(axis='y', labelsize=10)
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
Method = ['Jam then Cream', 'Cream then Jam', 'No Answer', 'Any Combination', 'Half & Half', 'No Scone!', 'Only Jam', 'Only Cream', 'Cheese']
Count = [8, 6, 42, 7, 4, 4, 2, 1, 1]

# Create pie chart
plt.pie(Count, labels=Method, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('The Scone Debate')

# Show the plot
plt.show()
```
