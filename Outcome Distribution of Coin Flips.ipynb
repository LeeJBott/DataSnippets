```python
import random
import matplotlib.pyplot as plt
```

```python
def coin():
    return random.randint(0, 1)
```

```python
def outcome(result):
    if result == 0:
        print("Heads")
    else:
        print("Tails")
```

```python
def singFlip():
    result = coin()
    outcome(result)
```

```python
def multFlip(num):
    for i in range(num):
        result = coin()
        outcome(result)
        list1.append(result)
    print("Data outcome of your Flips!:")
    print(list1)
```

```python
if __name__ == "__main__":
    list1 = []    
    ask1 = input("Flip (O)nce or (M)ultiple?: ")
    if ask1.lower() == "o":
        singFlip()
    elif ask1.lower() == "m":
        ask2 = int(input("How many flips?: "))
        multFlip(ask2)
        
        heads_count = list1.count(0)
        tails_count = list1.count(1)

        labels = ['Heads', 'Tails']
        counts = [heads_count, tails_count]
        #bar graph
        plt.bar(labels, counts)
        plt.title('Outcome Distribution of Coin Flips')
        plt.xlabel('Outcome')
        plt.ylabel('Count')
    
        plt.show()
```
